```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

#### Testing the differences between MAPE and SMAPE.  
Both versions of SMAPE will be used (100 and 200).

```python
actual = np.array([100.0 for _ in range(0,10000)])
forecast = np.array([0 + i/100 for i in range(0, len(actual))]) * actual

print("Actual is always 100.0. \n\nForecast is:", *forecast[:5], '...',
      *forecast[-5:])
```

```python
percentage_error = (100*((forecast-actual) / actual))

ape = (100*abs((actual-forecast) / actual))
sape100_1 = (100*abs(actual-forecast)) / (abs(actual+forecast))
sape200_1 = (100*abs(actual-forecast)) / ((abs(actual+forecast)/2))

plt.plot(percentage_error, ape)
plt.plot(percentage_error, sape100_1)
plt.plot(percentage_error, sape200_1)
plt.legend(['ape', 'sape100_1', 'sape200_1'], loc='upper left')
plt.xlim(-100, 100)
plt.ylim(-10, 250)
plt.show;
```

```python
print("Showing SMAPE upper bond for bigger differences:")
plt.plot(percentage_error, ape)
plt.plot(percentage_error, sape100_1)
plt.plot(percentage_error, sape200_1)
plt.legend(['ape', 'sape100_1', 'sape200_1'], loc='upper left')
plt.ylim(-10, 250)
plt.show;
```

```python
print("Another way to calculate SMAPEs. ",
      "Only differs if negative values are present (not the case here)")
sape100_2 = (100*abs(actual-forecast)) / (abs(actual)+abs(forecast))
sape200_2 = (100*abs(actual-forecast)) / ((abs(actual)+abs(forecast))/2)

plt.plot(percentage_error, ape)
plt.plot(percentage_error, sape100_2)
plt.plot(percentage_error, sape200_2)
plt.legend(['ape', 'sape100_2', 'sape200_2'], loc='upper left')
plt.xlim(-100, 100)
plt.ylim(-10, 250)
plt.show;
```
